```python
def find_max_column(df, col):
    max_col = None
    max_val = -1
    for c in col:
        if df[c].max() > max_val:
            max_col = c
            max_val = df[c].max()
    return (max_col, max_val)
```

```python
def get_bin_range(min_val, max_val, num_bins, bin_index):
    bin_width = (max_val - min_val) / num_bins
    lower = min_val + bin_index * bin_width
    upper = lower + bin_width
    return round(lower, 3), round(upper, 3)
```

```python
n_bins = {}
for bin in  nitrogen_c:
    vals = get_bin_range(0.02, 0.7, 7, int(bin.split("_")[2]))
    n_bins[bin] = str(vals[0]) + "-" + str(vals[1])
```

```python
p_bins = {}
for bin in  phosphorus_c:
    vals = get_bin_range(0.01, 0.64, 7, int(bin.split("_")[2]))
    p_bins[bin] = str(vals[0]) + "-" + str(vals[1])
```

```python
ph_bins = {}
for bin in  ph_c:
    vals = get_bin_range(3.42, 8.14, 5, int(bin.split("_")[2]))
    ph_bins[bin] = str(vals[0]) + "-" + str(vals[1])
```

```python
soil_stats["median_nitrogen_levels"] = soil_stats["median_nitrogen_levels"].replace(n_bins)
soil_stats["median_phosphorus_levels"] = soil_stats["median_phosphorus_levels"].replace(p_bins)
soil_stats["median_ph_levels"] = soil_stats["median_ph_levels"].replace(ph_bins)
soil_stats
```
