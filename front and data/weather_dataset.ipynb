```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("C:/Users/Dima/Desktop/MIP/stations/combined_daily_frame_with_seasons_to_be_aggr.csv")
dataset.info()
```

```python
dataset.groupby("Year").count()
```

```python
dataset.Season.unique()
```

```python
grouped_ds = dataset.groupby(["Region", 'Date'])["daily_rain"].sum().reset_index()
```

```python
grouped_ds.info()
```

```python
grouped_ds["radiation"] = dataset.groupby(["Region", 'Date'])["radiation"].sum().reset_index()["radiation"]
grouped_ds
```

```python
grouped_ds["max_temp"] = dataset.groupby(["Region", 'Date'])["max_temp"].max().reset_index()["max_temp"]
grouped_ds["min_temp"] = dataset.groupby(["Region", 'Date'])["min_temp"].min().reset_index()["min_temp"]
grouped_ds
```

```python
grouped_ds["rh_tmax"] = dataset.groupby(["Region", 'Date'])["rh_tmax"].mean().reset_index()["rh_tmax"]
grouped_ds["rh_tmin"] = dataset.groupby(["Region", 'Date'])["rh_tmin"].mean().reset_index()["rh_tmin"]
grouped_ds["et_short_crop"] = dataset.groupby(["Region", 'Date'])["et_short_crop"].mean().reset_index()["et_short_crop"]
grouped_ds
```

```python
aggregates = grouped_ds.merge(dataset[["Region", "Date", "Year", "Month", "Day", "Season"]].drop_duplicates(), on=["Region", "Date"], how = "left")
aggregates
```

```python
yields_data = pd.read_csv("C:/Users/Dima/Desktop/MIP/yield_data.csv")
yields_data[["Year", "Region", "Production"]]
```

```python
aggregates_full = aggregates.merge(yields_data[["Year", "Region", "Production"]], on = ["Year", "Region"], how = "left")
aggregates_full
```
